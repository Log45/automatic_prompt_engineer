```python
# First, let's define a simple dataset consisting of words and their labelels relevant to chemistry
words = ["hydrochloric acid", "90Sr/90Y", "D5041 cattion exchange resin", "0.2-0.3 M", 
         "HNO3", "HCl", "90Y", "45 days", "0.1%"]
labels = ["acid", "metal", "chemical compound", "quantity", "acid", 
          "acid", "metal", "time", "quantity"]
```

```python
# Now, we need to define the format of the prompt that we are using.

eval_template = \
"""Instruction: [PROMPT]
Input: [INPUT]
Output: [OUTPUT]"""
```

```python
# Now, let's use APE to find prompts that generate labels for each word.
from automatic_prompt_engineer import ape

result, demo_fn = ape.simple_ape(
    dataset=(words, labels),
    eval_template=eval_template,
    eval_model="facebook/opt-1.3b",
    prompt_gen_model="facebook/opt-1.3b",
    prompt_gen_mode="opt",
    prompt_gen_batch_size=100,
    eval_batch_size=250
)
```

```python
# Let's see the results.
print(result)
```

```python
from automatic_prompt_engineer import ape

manual_prompt = "Act like you are a professional chemist and label the following phrases."

human_result = ape.simple_eval(
    dataset=(words, labels),
    eval_template=eval_template,
    prompts=[manual_prompt],
    eval_model="facebook/opt-1.3b"
)
```
